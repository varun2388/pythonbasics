```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.figure(figsize =(4,4))
plt.hist(df1['Accept'])
plt.xlabel('Accept')
plt.ylabel('count')
plt.title('histgram of accept')
plt.show()
```

```python
plt.figure(figsize =(4,4))
plt.hist(df1['SAT'])
plt.xlabel('SAT')
plt.ylabel('count')
plt.title('histogram of SAT')
plt.show()
```

```python
sns.histplot(df1['Accept'])
```

```python
sns.histplot(df1['SAT'])
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(df1['Accept'])
plt.xlabel('Accept')
```

```python
plt.plot(df1['Accept'],df1['Expenses'], 'r+')
plt.xlabel('Accept')
plt.ylabel('Expenses')
plt.title('Scatterplot accept vs expenses')

help(plt.plot)
help(plt.hist)
```
